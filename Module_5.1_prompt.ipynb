### Required Assignment 5.1: Will the Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaurant near where you are driving. Would you accept that coupon and take a short detour to the restaurant? Would you accept the coupon but use it on a subsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaurant? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. 3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
# Define the groups based on monthly bar visits
infrequent_bar_goers = ['never', 'less1', '1~3', 'unknown']
frequent_bar_goers = ['4~8', 'gt8']

# Filter and calculate acceptance rate for those who go 3 or fewer times
rate_3_or_fewer = bar_data[bar_data['Bar'].isin(infrequent_bar_goers)]['Y'].mean()

# Filter and calculate acceptance rate for those who go more than 3 times
rate_more_than_3 = bar_data[bar_data['Bar'].isin(frequent_bar_goers)]['Y'].mean()

# Print the comparison results
print(f"Acceptance rate for those who visit bars 3 or fewer times/month: {rate_3_or_fewer * 100:.2f}%")
print(f"Acceptance rate for those who visit bars MORE than 3 times/month: {rate_more_than_3 * 100:.2f}%")
print(f"Difference: {(rate_more_than_3 - rate_3_or_fewer) * 100:.2f}%")
```

```python
# Observation:
# Drivers who rarely or never go to bars accept the coupon at a low rate (around 37%).
# Drivers who regularly visit bars (more than 3 times a month) accept the coupon at 
# a massive rate (around 76%).

# This is a classic validation of historical behavioral data over circumstantial data. 
# A driver's past habits are a massive predictor of their future coupon acceptance.
# Even though the overall acceptance rate for bar coupons is a weak 41.00%, 
# targeting frequent bar-goers yields a massive 76% conversion rate. 
# 
# Recommendation:
# The business should restrict bar coupon distribution heavily to individuals 
# with a proven history of visiting bars to avoid wasting advertising spend.
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
# 1: Need to create 2 groups: 
# Group A (The Target Group): Drivers whose monthly bar visits are more 
# than once ('1~3', '4~8', or 'gt8') AND whose age is over 25.

# Group B (All Others): Everyone else who received a bar coupon but doesn't 
# meet both of those conditions.

# 2: Age is defined as strings.  Must be handled appropriately
# ======================

# 1. Define the criteria for Group A
frequent_bar_categories = ['1~3', '4~8', 'gt8']
age_over_25_categories = ['26', '31', '36', '41', '46', '50plus']

# 2. Filter for Group A (Goes > 1 time/month AND age > 25)
group_A_condition = (bar_data['Bar'].isin(frequent_bar_categories)) & (bar_data['age'].isin(age_over_25_categories))
group_A_data = bar_data[group_A_condition]
rate_group_A = group_A_data['Y'].mean()

# 3. Filter for Group B (All others)
group_B_data = bar_data[~group_A_condition] # '~' == NOT, which means all others 
rate_group_B = group_B_data['Y'].mean()

# 4. Print the results
print(f"Acceptance rate for Group A (Goes > 1x/month & Age > 25): {rate_group_A * 100:.2f}%")
print(f"Acceptance rate for Group B (All Others): {rate_group_B * 100:.2f}%")
print(f"Absolute Difference: {abs(rate_group_A - rate_group_B) * 100:.2f}%")
```

```python
# Observations:
# There is a massive difference - Group A accepts the bar coupon at roughly 69.5%, 
# while Group B (All Others) accepts it at roughly 33.5%.

# By combining just two demographic factors—age and historical habit
# we can isolate a segment of the population that is more than twice as 
# likely to accept a bar coupon compared to the rest of the driving population. 
# It shows that while bar coupons are generally unpopular (with a low 41% baseline), 
# they become highly effective tools if young drivers (under 25) and those who 
# don't already have a habit of visiting bars at least a couple of times a month are filtered out.
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry.


```python
#First Inspect the data:
# We already know the bar-visit data type and values
# Verify expected values of passengers and occupation

print(data['passenger'].unique())
print(data['occupation'].unique())
```

```python
# Next we need to define the criteria for the target group (Group-A). Must meet all criteria below:
# a: Bar Habit: Goes to a bar more than once a month ('1~3', '4~8', or 'gt8').
# b: Passenger: Has a passenger that is not a kid, meaning the passenger is a 'Friend(s)' or 'Partner'.
#    We also want to exclude driving alone ('Alone'), as the prompt implies they had passengers.
# c: Occupation: Work in fields other than farming, fishing, or forestry ('Farming Fishing & Forestry').

# 1. Define the specific inclusion/exclusion lists
frequent_bar_categories = ['1~3', '4~8', 'gt8']
valid_passengers = ['Friend(s)', 'Partner']  # Has passengers, but not kids

# 2. Build the combined logical condition for Group A
group_A_condition = (
    (bar_data['Bar'].isin(frequent_bar_categories)) &
    (bar_data['passenger'].isin(valid_passengers)) &
    (bar_data['occupation'] != 'Farming Fishing & Forestry')
)

# 3. Split the data and calculate acceptance rates
group_A_data = bar_data[group_A_condition]
rate_group_A = group_A_data['Y'].mean()

group_B_data = bar_data[~group_A_condition] # all others
rate_group_B = group_B_data['Y'].mean()

# 4. Print the comparison results
print(f"Total drivers in Group-A, our target: {len(group_A_data)}")
print(f"Acceptance rate for Group-A (Frequent bar-goers, with adult peers, non-farming): {rate_group_A * 100:.2f}%")
print(f"Acceptance rate for Group-B (All Others): {rate_group_B * 100:.2f}%")
print(f"Absolute Difference: {abs(rate_group_A - rate_group_B) * 100:.2f}%")
```

```python
#Observation:
# Group A accepts the coupon at a massive rate of roughly 71.3%, 
# while Group B (All Others) sits around 37.8%.


# This condition perfectly zeroes in on the social driver. They already like going to bars (Bar > 1).
# They are currently in the car with their peers or significant others (Friend(s) or Partner), 
# meaning they are likely already out for a social event or open to starting one. They don't 
# have children in the car to worry about getting home to, and they don't work
# in highly localized/outdoor primary industries like forestry or 
# farming where work hours and locations look very different from typical commercial bar districts.

# This insight can be used to target the correct audience
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K.



```python
#We need to connect three separate profiles using OR. 
# This means if a driver fits Profile 1, Profile 2, OR Profile 3, they belong in our target group 
# (Group A). Everyone else falls into Group B.

#The profiles
# Profile 1: Goes to bars > 1x/month ('1~3', '4~8', 'gt8') 
#            AND passengers are not kids ('Friend(s)', 'Partner') 
#            AND marital status is not widowed ('Widowed').

# Profile 2: Goes to bars > 1x/month ('1~3', '4~8', 'gt8') 
#         AND age is under 30 ('below21', '21', '26').

# Profile 3: Goes to cheap restaurants > 4x/month ('4~8', 'gt8') 
#         AND income is under 50K ('Less rather than $12500', 
#              '$12500 - $24999', '$25000 - $37499', '$37499 - $49999').
```

```python
# let us first inspect the data:
print(bar_data['maritalStatus'].unique())
print(bar_data['income'].unique())
print(bar_data['RestaurantLessThan20'].unique())
```

```python
#Next let us calculate the acceptance rate:

# --- Define Categorical Groups ---
frequent_bars = ['1~3', '4~8', 'gt8']
no_kids = ['Friend(s)', 'Partner']
under_30 = ['below21', '21', '26']
frequent_cheap_rest = ['4~8', 'gt8']

low_income = [
    'Less rather than $12500', 
    '$12500 - $24999', 
    '$25000 - $37499', 
    '$37499 - $49999'
]

# --- Build the 3 Distinct Profiles ---
profile_1 = (
    (bar_data['Bar'].isin(frequent_bars)) & 
    (bar_data['passenger'].isin(no_kids)) & 
    (bar_data['maritalStatus'] != 'Widowed')
)

profile_2 = (
    (bar_data['Bar'].isin(frequent_bars)) & 
    (bar_data['age'].isin(under_30))
)

profile_3 = (
    (bar_data['RestaurantLessThan20'].isin(frequent_cheap_rest)) & 
    (bar_data['income'].isin(low_income))
)

# --- Combine Profiles using OR (|) ---
group_A_condition = profile_1 | profile_2 | profile_3

# --- Calculate Acceptance Rates ---
group_A_data = bar_data[group_A_condition]
rate_group_A = group_A_data['Y'].mean()

group_B_data = bar_data[~group_A_condition]
rate_group_B = group_B_data['Y'].mean()

# --- Print the Results ---
print(f"Total drivers in Group-A, the target: {len(group_A_data)}")
print(f"Acceptance rate for Group A (Matches any of the 3 profiles): {rate_group_A * 100:.2f}%")
print(f"Acceptance rate for Group B (All Others): {rate_group_B * 100:.2f}%")
print(f"Absolute Difference: {abs(rate_group_A - rate_group_B) * 100:.2f}%")
```

```python
# Looking a passengers under 30, with a friend and time of day


# Define our target demographics
young_age = ['below21', '21', '26']
coffee_times = ['7AM', '10AM', '2PM']

# Build the composite filter
coffee_profile_condition = (
    (coffee_data['passenger'] == 'Friend(s)') &
    (coffee_data['age'].isin(young_age)) &
    (coffee_data['time'].isin(coffee_times))
)

# Calculate rates
target_coffee_group = coffee_data[coffee_profile_condition]
rate_target_coffee = target_coffee_group['Y'].mean()
rate_coffee_others = coffee_data[~coffee_profile_condition]['Y'].mean()

print(f"Total drivers matching this ideal coffee profile: {len(target_coffee_group)}")
print(f"Acceptance rate for Target Coffee Profile: {rate_target_coffee * 100:.2f}%")
print(f"Acceptance rate for All Other Drivers: {rate_coffee_others * 100:.2f}%")
```

```python
#What is I look at "Alone" with all other conditions same?

young_age = ['below21', '21', '26']
coffee_times = ['7AM', '10AM', '2PM']

# Build the composite filter for ALONE drivers
alone_coffee_condition = (
    (coffee_data['passenger'] == 'Alone') &
    (coffee_data['age'].isin(young_age)) &
    (coffee_data['time'].isin(coffee_times))
)

# Calculate rates
alone_coffee_group = coffee_data[alone_coffee_condition]
rate_alone_coffee = alone_coffee_group['Y'].mean()

print(f"Total drivers matching the ALONE profile: {len(alone_coffee_group)}")
print(f"Acceptance rate for young, midday drivers ALONE: {rate_alone_coffee * 100:.2f}%")
```
